```python
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/AAISOL/ai4i2020.csv")
#df.info()
df
```

```python
#Columns with Kelvin temperatures
AirTemp_K = 'Air temperature [K]'
ProcessTemp_K = 'Process temperature [K]'

#Columns with RPM
rpm_column = 'Rotational speed [rpm]'

#Mapping for Type column
replace_mapping = {'L': 0, 'M': 1, 'H' : 3}

#Replace Kelvin with Celsius values
df[AirTemp_K] = df[AirTemp_K] - 273.15
df[ProcessTemp_K] = df[ProcessTemp_K] - 273.15

#RPM to RPS
df[rpm_column] = df[rpm_column] / 60

#Update column name
new_column_name = 'Air temperature [C]'
df.rename(columns={AirTemp_K: new_column_name}, inplace=True)
new_column_name = 'Process temperature [C]'
df.rename(columns={ProcessTemp_K: new_column_name}, inplace=True)
new_column_name = 'Rotational speed [rps]'
df.rename(columns={rpm_column: new_column_name}, inplace=True)

#Deleted not needed column
df.drop(columns='UDI', inplace=True)

#Use mapping for Type column
df['Type'] = df['Type'].replace(replace_mapping)

#Check if any failure have occured
if 1 not in df['Machine failure'].values:
    #if not delete the reasons column
    df.drop(columns=['TWF', 'HDF', 'PWF', 'OSF', 'RNF'], inplace=True)

df
```
